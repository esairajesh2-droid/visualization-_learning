```python
#imorting the required libraries
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({""
"Hourse_studied":[2,4,6],
"Marks":[40,60,80],
"Sleeo":[9,8,7]})
print(df)
corr=df.corr()
print(corr)
plt.figure(figsize=(8,5))
plt.title('Heatmap visualization',color='Skyblue',fontsize=25)
plt.imshow(corr,cmap='coolwarm')
plt.xticks(range(len(corr.columns)),corr.columns)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.colorbar(label="correlation")
plt.show()
```

```python
#fitness dataset analysis
df = pd.DataFrame({
    "Exercise_Hours": [1, 2, 3, 4, 5],
    "Calories_Burned": [200, 350, 500, 650, 800],
    "Weight": [80, 78, 76, 74, 72]
})
corr=df.corr()
print(corr)
plt.style.use("Solarize_Light2")
plt.figure(figsize=(8,5))
plt.title('fitness analysis',color='skyblue',fontsize=25)
plt.imshow(corr)
plt.xticks(range(len(corr.columns)),corr.columns)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.colorbar(label="correlation")
plt.show()
```

```python
#Exercise Hours and Calories Burned have a strong positive correlation. As exercise hours increase, calories burned also increase. Exercise Hours and Weight have a
#strong negative correlation, so weight tends to decrease as exercise hours increase.
```

```python
#employee dataset heatmap
df = pd.DataFrame({
    "Experience": [1, 3, 5, 7, 9],
    "Salary": [25000, 35000, 50000, 70000, 90000],
    "Stress_Level": [9, 8, 6, 5, 4]
})
corr=df.corr()
plt.style.use("dark_background")
plt.title("employee heatmap",color="blue",fontsize=25)
plt.imshow(corr,cmap="coolwarm")
plt.xticks(range(len(corr.columns)),corr.columns)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.colorbar()
plt.show()
```

```python
#cricket heatmap
df = pd.DataFrame({
    "Matches": [10, 20, 30, 40, 50],
    "Runs": [350, 700, 1050, 1400, 1750],
    "Strike_Rate": [120, 125, 130, 135, 140]
})
corr=df.corr()
print(corr)
plt.style.use("dark_background")
plt.title("cricket heatmap",color='blue',fontsize=25)
plt.imshow(corr,cmap="Blues")
plt.xticks(range(len(corr.columns)),corr.columns,rotation=15)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.colorbar(label="correlation")

plt.show()
```

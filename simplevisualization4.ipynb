```python
#importing the required 
import matplotlib.pyplot as plt
import seaborn as sns
students = ["A", "B", "C", "D", "E", "F", "G"]
marks = [45, 60, 72, 88, 91, 76, 55]
plt.title("Stduent Marks",fontsize=25,color="red")
sns.barplot(x=students,y=marks,orient="v",hue=students)
plt.xlabel("Studnets",fontsize=25,labelpad=8,color="red")
plt.ylabel("Marks",fontsize=25,labelpad=8,color="red")
plt.xticks(color="black",fontsize=15)
plt.yticks(color="black",fontsize=15)
plt.gca().set_facecolor("white")
plt.tick_params()
plt.gcf().set_facecolor("skyblue")
plt.show()
```

```python
#impoting the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\esair\Downloads\retail_sales_dataset.csv")
print(df)
```

```python
#statistical data of the dataset
df.describe()
```

```python
#diving the columns in categorical and numerical
df_categorical=df.select_dtypes(include=['object','str']).columns
print('categorical columns')
print(df_categorical)
```

```python
df_numerical=df.select_dtypes(include='number').columns
print(df_numerical)
```

```python
#who is person with total amount spent with age in top 100
sample=df.head(100)
sns.scatterplot(data=sample,x='Age',y='Total Amount',hue='Gender',s=50,alpha=0.5,palette='Set1')
plt.gca().set_facecolor('white')
plt.gcf().set_facecolor('skyblue')
plt.show()
```

```python
#which age has spent more mmoney 
sns.lineplot(data=sample,x='Age',y='Total Amount',alpha=0.5)
```

```python
#now learning lineplot for predicting daily sales 
print(df)
df.Date=pd.to_datetime(df.Date)
daily_sales=df.groupby(['Date','Gender'])['Total Amount'].sum().reset_index()
plt.figure(figsize=(12,5))
sns.lineplot(data=daily_sales,x='Date',y='Total Amount',hue=df.Gender)
plt.xticks(rotation=45)
plt.show()
```

```python
#barplot using seaborn
bar=df.groupby(df.Gender)["Age"].mean().reset_index()
sns.barplot(data=bar,x='Gender',y='Age',errorbar=None)
plt.show()
```

```python
sales=df.groupby("Product")["Total Amount"].sum().reset_index()
sns.countplot(data=sales,x="Product",y="Total Amount")
```

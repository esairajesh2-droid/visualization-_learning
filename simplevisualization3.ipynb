```python
#importing required libraries
import matplotlib.pyplot as plt
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug"]

sales = [120,150,180,170,210,250,280,300]
plt.figure(figsize=(8,5))
plt.style.use("ggplot")
plt.title("Monthly Sales",color="blue")
plt.plot(months,sales,color="red")
plt.fill_between(months,sales,alpha=0.8,color="yellow")
plt.xlabel("months")
plt.ylabel("sales")
plt.show()
```

```python
days = [1,2,3,4,5,6,7]

temperature = [31,32,34,33,35,36,34]
plt.figure(figsize=(8,5))
plt.style.use("ggplot")
plt.title("Daily Temperatur",color="black")
plt.plot(days,temperature,color="blue")
plt.fill_between(days,temperature,alpha=0.8,color="yellow")
plt.xlabel("days")
plt.ylabel("temperatur")
plt.show()
```

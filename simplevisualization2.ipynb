```python
#importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({'months':["Jan","Feb","Mar","Apr","May","Jun"],
'mobile':[30,40,35,50,60,70],
'laptop':[20,25,30,35,40,45],
'tablet':[10,12,15,18,20,25]
})
plt.figure(figsize=(9,5))
plt.stackplot(df.months,df.mobile,df.laptop,df.tablet,colors=['red','green','yellow'],alpha=0.8,labels=["mobiles","laptops","tablet"])
plt.title("Company Sales Contribution")
plt.xlabel("Months")
plt.ylabel("Units Sold")
plt.legend(loc="upper left")
plt.show()
```

```python
months=["Jan","Feb","Mar","Apr","May"]

math=[50,55,60,62,65]

science=[45,50,55,58,60]

english=[40,42,45,48,50]
plt.figure(figsize=(9,5))
plt.stackplot(months,math,science,english,labels=["math","science","english"],alpha=0.6)
plt.title("Marks stackplot")
plt.xlabel("Months")
plt.ylabel("Marks")
plt.legend(loc="upper left")
plt.show()
```

```python
matches=[1,2,3,4,5]

goals=[1,2,0,3,1]

assists=[0,1,2,1,1]

tackles=[4,5,3,6,4]
plt.figure(figsize=(9,5))
plt.gcf().set_facecolor("skyblue")
plt.stackplot(matches,goals,assists,tackles,labels=["goals","assists","tackles"],alpha=0.8)
plt.title("Football Stackplot")
plt.xlabel("Matches")
plt.ylabel("Scores")
plt.legend(loc="upper right")
plt.show()
```
